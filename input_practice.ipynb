### 2.1 [실습 - 숫자 맞히기 게임](https://www.codeit.kr/topics/making-use-of-python/lessons/2821)

- `while (조건문)`: 

    - 여기서 조건문은 조건문만 만족한다면 무한 반복하겠다는 의미입니다!

- 버전1 : 목숨(LIFES)만 기준으로 while문 반복 실행

````python
import random

LIFES = 4 # 시도 횟수 제한
ANSWER = random.randint(1,20) # 1~20 사이의 랜덤 정수 생성

guess = -1 # 사용자가 입력한 숫자 초기화
tries = 0 # 시도 횟수

while (guess != ANSWER) and (tries < LIFES):

    guess = int(input (f"기회가 {LIFES}번 남았습니다. 1~20 사이의 숫자를 맞혀보세요: "))

    tries += 1

    if guess < ANSWER:
        print("UP")
    elif guess > ANSWER:
        print("DOWN")

if guess == ANSWER:
    print(f"축하합니다! {tries}번 만에 맞히셨습니다.")
else:
    print(f"아쉽습니다. 정답은 {ANSWER}였습니다.")
````
