### 3.3 [실습 - 코딩에 빠진 닭](https://www.codeit.kr/topics/making-use-of-python/lessons/2825)

- data/chichen.txt 에는 치킨집(코딩의 빠진 닭)의 12월 매출이 정리되어 있습니다.

    - 일별 매출 값을 읽어와서, 하루 평균 매출을 구하세요.

- 버전 1: 변수명 꼼꼼히 정의해주는 정석

````python
with open ('./data/chicken.txt', 'r', encoding='utf-8') as f: #불러올 자료
    
    total_revenue = 0 #총 매출
    total_days = 0   #영업일 수

    for line in f: #한 줄씩 읽기
        revenue = line.strip().split(': ')[1] #날짜 뒤의 매출액 부분만 출력

        total_revenue += int(revenue)  #매출액을 총 매출에 더하기

        total_days += 1          #영업일 수 1 증가
    
    print("총 매출:", total_revenue / total_days) #평균 매출 출력


    
````

- 버전 2: 가장 간결하게 `메소드 체이닝(method chaining)`으로 작성

### 3.5 [실습 - 단어장 만들기](https://www.codeit.kr/topics/making-use-of-python/lessons/2827)

- "영어: 한국어" 이런식으로 단어장(vocabulary.txt)에 정리하도록 하는 프로그램을 만들어볼게요!

- 2가지의 입력을 받습니다.

    ```bash
    - 영어 단어를 입력하세요:
    - 한국어 뜻을 입력하세요:
    # 위 둘 중에 q를 넣으면 바로 프로그램 종료 
    ```

- 버전 1: While을 바깥 껍질에

````python
while True: #조건문이 false가 될 수 없으니 무한 반복

    eng_vocab = input("영어 단어를 입력하세요(종료하려면 q 입력): ")
    if eng_vocab == 'q': #만약 q를 입력하면 반복문 종료
        print("단어장 작성을 종료합니다.")
        break   
    kor_vocab = input("한국어 단어를 입력하세요(종료하려면 q 입력): ")
    if kor_vocab == 'q': #만약 q를 입력하면 반복문 종료
        print("단어장 작성을 종료합니다.")
        break 

    with open ('./data/vocabulary.txt', 'a', encoding='utf-8') as f: #단어장 파일 열기(추가 모드)
        f.write(f"{eng_vocab}: {kor_vocab}\n") #입력한 단어를 파일에 쓰기

print("단어장이 성공적으로 작성되었습니다.") 

````

- 버전2 : `with open() as f:` 를 바깥 껍질에

````python
with open ('./data/vocabulary.txt', 'a', encoding='utf-8') as f:
    eng_vocab = input("영어 단어를 입력하세요: ")
    kor_vocab = input("한국어 단어를 입력하세요: ")
    # 위 둘 중에 q를 넣으면 바로 프로그램 종료 

    while True:
        if eng_vocab == 'q': # q 입력 시 프로그램 종료
            break
        if kor_vocab == 'q': # q 입력 시 프로그램 종료
            break

    f.write(f"{eng_vocab}: {kor_vocab}\n") # 단어장에 단어 추가
````

- (틈새 개념) python의 2가지 포매팅 방법

    - 포매팅 : 지정된 자리에 값을 대입한다.

| **포인트**           | **f-string(권장)**                               | **str.format()**                              |
|----------------------|--------------------------------------------|------------------------------------------------|
| **Python 버전**       | Python 3.6 이상                           | Python 2.7, 3.x                               |
| **가독성**           | 간결하고 직관적                             | 다소 복잡함                                    |
| **표현식 사용**      | 가능 (중괄호 안에 표현식)                    | 불가능 (변수만 삽입 가능)                      |
| **성능**             | 더 빠름 (컴파일 단계에서 처리)                | 상대적으로 느림 (런타임에 처리)                |
| **사용성**           | 최신 문법으로 권장                           | 하위 호환성을 위해 여전히 사용 가능            |

### 3.6 [실습 - 단어 퀴즈](https://www.codeit.kr/topics/making-use-of-python/lessons/2828)

- 위에서 만든 vocabulary.txt를 바탕으로

    - 문제를 내는 프로그램을 만들어 봅시다!

    - 순서대로 불러오면서, 사용자가 영어단어를 입력하고,

        - 정답이면, `맞았습니다!`
        - 오답이면, `아쉽습니다. 정답은 OOO입니다.`가 되도록 해봅시다!

````python
with open ('./data/vocabulary.txt', 'r', encoding='utf-8') as f:
    for line in f:
        
        
        data = line.strip().split(": ")
        # print(data)
        eng_vocab, kor_vocab = data[0], data[1]  #뭉탱이를 strip으로 변환 split으로 해체

        #유저에게 문제 내기
        guess = input(f"{kor_vocab}의 영어 단어는 무엇일까요? (q 입력시 종료됩니다.)")

        if guess == 'q': # q 입력 시 프로그램 종료
            print("퀴즈를 종료합니다.")
            break
        
        if guess == eng_vocab:
            print("정답입니다!")
        else:
            print(f"아쉽습니다. 정답은 {eng_vocab}입니다.")
        
   
````

### 3.7 [실습 - 고급 단어장](https://www.codeit.kr/topics/making-use-of-python/lessons/2829)

- 위에서 만든 단어장을 바탕으로 학생들에 문제를 출제하는 프로그램
- `한국어: (정답)` 이런식으로 물어봅니다!

- 틀리면 `"아쉽습니다. 정답은 {영어}입니다."` 로 말합니다.

    - 3.6이랑 동일한데, **random하게 물어보도록**

    - 물론, **중복해서 질문** 하게될 수도 있어요!
    
        - hint. random이랑 dictionary를 이용할 것!

- 버전 1: 직관적으로 쉽게 짜기

> 1. 단어장 불러오기
>
>    - 우선 랜덤하게 보여주려면, 단어장 data를 모두 불러와서 dict에 기록해놓을 필요가 있어요!

````python
with open ('./vocabulary.txt', 'r', encoding='utf-8') as f:

    ko_en_word_dict = {} #빈 딕셔너리 생성
    
    for line in f:
        
        data = line.strip().split(": ") #화이트 스페이스, 줄넘김 제거
        en, ko = data[0], data[1]  #언패킹 기법 사용

        ko_en_word_dict[ko] = en  #딕셔너리에 단어 추가

ko_en_word_dict
        
````

> 2. 랜덤하게 출제하기
>
>    - 이제 random 함수로 dict의 항들을 임의로 꺼내서, 문제를 내봅시다!

````python
list(ko_en_word_dict.keys())[5]
````

````python
list(ko_en_word_dict.values())[5]
````

````python
import random #랜덤 모듈 불러오기
    
while True:

    rand_idx = random.randint(0, len(ko_en_word_dict)-1) #딕셔너리 인덱스 범위 내에서 랜덤 정수 생성
    
    # 3번 방식
    kor_vocab = list(ko_en_word_dict.keys())[rand_idx] 
    eng_vocab = list(ko_en_word_dict.values())[rand_idx]

    #유저에게 문제 내기
    guess = input(f"{kor_vocab}의 영어 단어는 무엇일까요?(q 입력시 종료) ")

    if guess == 'q': # q 입력 시 프로그램 종료
        print("퀴즈를 종료합니다.")
        break
    
    if guess == eng_vocab:
        print("정답입니다!")
    else:
        print(f"아쉽습니다. 정답은 {eng_vocab}입니다.")
````
