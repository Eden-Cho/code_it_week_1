### 1.1 모듈 불러오기

#### - 권장 1

>    - aliasing으로 모듈의 줄임말 닉네임을 붙여서
>
>    - .(func) 으로 기능(메서드, 함수)을 호출해서 사용하기

````python
import calculator as calc

calc.add(2, 8) #덧셈

calc.subtract(10, 4) #뺄셈

calc.multiply(3, 7) #곱셈

calc.divide(20, 5) #나눗셈
````

#### 1. math
>   - 수학 관련 기능들의 묶음

````python
import math

print(math.log10(100)) #로그 함수
print(math.cos(0))   #코사인 함수
print(math.pi)  #원주율 상수
````

#### 2. random
>   - 랜덤한 값을 만들어내는 기능

````python
import random

print(random.random()) #0.0에서 1.0 사이의 임의의 부동 소수점 숫자 생성

print(random.randint(1, 20)) #1에서 20 사이의 임의의 정수 생성

print(random.uniform(0, 2)) #0.0에서 2.0 사이의 임의의 부동 소수점 숫자 생성
````

#### 3. os
>   - 운영체제(operating system)를 조작하는 기능들의 묶음

````python
import os #운영체제 조작 기능

print(os.getlogin())  #현재 로그인한 사용자의 이름 반환

print(os.getcwd())    #현재 작업 디렉토리 반환
````

#### 4. datetime
>   - 날짜와 시간을 다루기 위한 다양한 기능들의 묶음

````python
import datetime
````

4.1 datetime 값 생성
- 2024년 9월 6일을 파이썬으로 어떻게 표현하면 될까요?


````python
today = datetime.datetime(2024, 9, 6)
print(today)
print(type(today))
````

- 시간은 자동으로 00시 00분 00초로 되어있는데, 이번엔 시간까지도 직접 정해볼까요?

````python
today = datetime.datetime(2024, 9, 6, 13, 13, 29)
print(today)
print(type(today))
````

4.3 timedelta 타입

- 날짜 사이의 차이를 나타내는 타입
    - delta: 수학에서 변화량을 나타내는 그리스어 기호

- 시간끼리 사칙연산이 가능할까요? 
    
    - 곱하기/나누기는 당연히 안되겠죠, 그래도 덧셈/ 뺄셈은 가능하지 않을까요?

````python
today_auto = datetime.datetime.now()
print(today_auto)

print(today_auto - today)
print(type(today_auto - today))
````

````python
print(today_auto + today)
````

````python
timedelta = datetime.timedelta(days=2, hours=3, minutes=10, seconds=12)

print(today_auto + timedelta)
````

4.4 datetime 해부하기
- datetime을 구성하는 연, 월, 일, 시, 분, 초를 추출하려면?

````python
print(today_auto)
````

````python
print(today_auto.year)  #현재 연도 반환
print(today_auto.month) #현재 월 반환
print(today_auto.day)   #현재 일 반환
print(today_auto.hour)  #현재 시 반환
print(today_auto.minute) #현재 분 반환
print(today_auto.second) #현재 초 반환
print(today_auto.microsecond) #현재 마이크로초 반환
````

4.5 datetime 포매팅
- datetime 값을 있는 그대로 출력하면 별로 예쁘지 않아요. 그러나 strftime() 함수를 사용하면 우리 입맛대로 바꿀 수 있어요!

    | 포맷 코드 | 설명 | 예시 |
    |------------|------------------------------|--------|
    | %a         | 요일 (짧은 버전)             | Mon    |
    | %A         | 요일 (풀 버전)               | Monday |
    | %w         | 요일 (숫자 버전, 0~6, 0이 일요일) | 5      |
    | %d         | 일 (01~31)                   | 23     |
    | %b         | 월 (짧은 버전)               | Nov    |
    | %B         | 월 (풀 버전)                 | November |
    | %m         | 월 (숫자 버전, 01~12)        | 10     |
    | %y         | 연도 (짧은 버전)             | 16     |
    | %Y         | 연도 (풀 버전)               | 2016   |
    | %H         | 시간 (00~23)                 | 14     |
    | %I         | 시간 (00~12)                 | 10     |
    | %p         | AM/PM                        | AM     |
    | %M         | 분 (00~59)                   | 34     |
    | %S         | 초 (00~59)                   | 12     |
    | %f         | 마이크로초 (000000~999999)    | 413215 |
    | %Z         | 표준시간대                   | PST    |
    | %j         | 1년 중 며칠째인지 (001~366)  | 162    |
    | %U         | 1년 중 몇 주째인지 (00~53, 일요일이 한 주의 시작이라고 가정) | 35 |
    | %W         | 1년 중 몇 주째인지 (00~53, 월요일이 한 주의 시작이라고 가정) | 35 |
